```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
try:
    from hdbscan import HDBSCAN
    HDBSCAN_AVAILABLE = True
except ImportError:
    HDBSCAN_AVAILABLE = False
    print("⚠️  HDBSCAN not installed. Only KMeans clustering will be available.")
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import umap
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re
import warnings
warnings.filterwarnings('ignore')

# Download required NLTK data
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
    
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')
    
try:
    nltk.data.find('corpora/wordnet')
except LookupError:
    nltk.download('wordnet')

class SemanticQuestionClustering:
    def __init__(self, model_name='all-MiniLM-L6-v2'):
        """
        Initialize the clustering system with BERT-based sentence transformer
        """
        print(f"🔄 Loading BERT model: {model_name}")
        self.model = SentenceTransformer(model_name)
        self.lemmatizer = WordNetLemmatizer()
        self.stop_words = set(stopwords.words('english'))
        
        # Results storage
        self.embeddings = None
        self.umap_embeddings = None
        self.cluster_labels = None
        self.silhouette_scores = {}
        
    def preprocess_text(self, text):
        """
        Comprehensive text preprocessing
        """
        if pd.isna(text):
            return ""
            
        # Convert to lowercase
        text = text.lower()
        
        # Remove special characters and digits (keep basic punctuation for context)
        text = re.sub(r'[^\w\s.,!?]', ' ', text)
        
        # Tokenize
        tokens = word_tokenize(text)
        
        # Remove stopwords and lemmatize
        processed_tokens = []
        for token in tokens:
            if token not in self.stop_words and len(token) > 2:
                lemmatized = self.lemmatizer.lemmatize(token)
                processed_tokens.append(lemmatized)
        
        return ' '.join(processed_tokens)
    
    def load_and_preprocess_data(self, csv_path):
        """
        Load CSV data and preprocess questions
        """
        print("📁 Loading and preprocessing data...")
        
        # Load data
        self.df = pd.read_csv(csv_path)
        print(f"✅ Loaded {len(self.df)} questions")
        
        # Check required columns
        if 'Questions' not in self.df.columns or 'Category' not in self.df.columns:
            raise ValueError("CSV must contain 'Questions' and 'Category' columns")
        
        # Preprocess questions
        print("🔄 Preprocessing text...")
        self.df['Processed_Questions'] = self.df['Questions'].apply(self.preprocess_text)
        
        # Remove empty questions after preprocessing
        self.df = self.df[self.df['Processed_Questions'].str.len() > 0].reset_index(drop=True)
        print(f"✅ {len(self.df)} questions after preprocessing")
        
        return self.df
    
    def generate_embeddings(self):
        """
        Generate BERT embeddings for preprocessed questions
        """
        print("🤖 Generating BERT embeddings...")
        
        # Use original questions for better semantic understanding
        # (preprocessing helps, but original text has more context)
        questions_for_embedding = self.df['Questions'].tolist()
        
        # Generate embeddings
        self.embeddings = self.model.encode(
            questions_for_embedding,
            show_progress_bar=True,
            batch_size=32
        )
        
        print(f"✅ Generated embeddings: {self.embeddings.shape}")
        return self.embeddings
    
    def reduce_dimensions(self, n_components=2, n_neighbors=15, min_dist=0.1):
        """
        Use UMAP for dimensionality reduction
        """
        print("📉 Reducing dimensions with UMAP...")
        
        reducer = umap.UMAP(
            n_components=n_components,
            n_neighbors=n_neighbors,
            min_dist=min_dist,
            metric='cosine',
            random_state=42
        )
        
        self.umap_embeddings = reducer.fit_transform(self.embeddings)
        print(f"✅ UMAP reduction complete: {self.umap_embeddings.shape}")
        
        return self.umap_embeddings
    
    def find_optimal_kmeans_clusters(self, k_range=range(5, 21)):
        """
        Find optimal number of clusters using Silhouette Score
        """
        print("🎯 Finding optimal number of clusters...")
        
        silhouette_scores = {}
        
        for k in k_range:
            kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
            cluster_labels = kmeans.fit_predict(self.embeddings)
            
            silhouette_avg = silhouette_score(self.embeddings, cluster_labels)
            silhouette_scores[k] = silhouette_avg
            
            print(f"K={k}: Silhouette Score = {silhouette_avg:.4f}")
        
        self.silhouette_scores = silhouette_scores
        
        # Find optimal k
        optimal_k = max(silhouette_scores, key=silhouette_scores.get)
        print(f"🏆 Optimal number of clusters: {optimal_k} (Score: {silhouette_scores[optimal_k]:.4f})")
        
        return optimal_k
    
    def perform_clustering(self, method='kmeans', n_clusters=None):
        """
        Perform clustering using KMeans or HDBSCAN
        """
        print(f"🔄 Performing {method.upper()} clustering...")
        
        if method == 'kmeans':
            if n_clusters is None:
                n_clusters = self.find_optimal_kmeans_clusters()
            
            clusterer = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
            self.cluster_labels = clusterer.fit_predict(self.embeddings)
            
        elif method == 'hdbscan':
            if not HDBSCAN_AVAILABLE:
                print("❌ HDBSCAN not available. Using KMeans instead...")
                return self.perform_clustering('kmeans', n_clusters)
                
            clusterer = HDBSCAN(
                min_cluster_size=5,
                min_samples=3,
                metric='cosine'
            )
            self.cluster_labels = clusterer.fit_predict(self.embeddings)
            
            # Handle noise points (-1) by assigning them to nearest cluster
            unique_clusters = np.unique(self.cluster_labels[self.cluster_labels >= 0])
            n_clusters = len(unique_clusters)
            
        else:
            raise ValueError("Method must be 'kmeans' or 'hdbscan'")
        
        # Add cluster labels to dataframe
        self.df['Cluster_Label'] = self.cluster_labels
        
        n_clusters_actual = len(np.unique(self.cluster_labels))
        print(f"✅ Clustering complete: {n_clusters_actual} clusters found")
        
        return self.cluster_labels
    
    def visualize_results(self, figsize=(15, 12)):
        """
        Create comprehensive visualizations
        """
        fig, axes = plt.subplots(2, 2, figsize=figsize)
        fig.suptitle('📊 Semantic Question Clustering Results', fontsize=16, fontweight='bold')
        
        # 1. Cluster visualization
        ax1 = axes[0, 0]
        scatter = ax1.scatter(
            self.umap_embeddings[:, 0], 
            self.umap_embeddings[:, 1],
            c=self.cluster_labels, 
            cmap='tab10', 
            alpha=0.7,
            s=50
        )
        ax1.set_title('🎯 Question Clusters (UMAP Projection)')
        ax1.set_xlabel('UMAP Component 1')
        ax1.set_ylabel('UMAP Component 2')
        plt.colorbar(scatter, ax=ax1, label='Cluster')
        
        # 2. Silhouette scores
        ax2 = axes[0, 1]
        if self.silhouette_scores:
            k_values = list(self.silhouette_scores.keys())
            scores = list(self.silhouette_scores.values())
            
            ax2.plot(k_values, scores, 'bo-', linewidth=2, markersize=8)
            ax2.set_title('📈 Silhouette Score vs Number of Clusters')
            ax2.set_xlabel('Number of Clusters (k)')
            ax2.set_ylabel('Silhouette Score')
            ax2.grid(True, alpha=0.3)
            
            # Highlight optimal k
            optimal_k = max(self.silhouette_scores, key=self.silhouette_scores.get)
            optimal_score = self.silhouette_scores[optimal_k]
            ax2.scatter([optimal_k], [optimal_score], color='red', s=100, zorder=5)
            ax2.annotate(f'Optimal k={optimal_k}\nScore={optimal_score:.3f}', 
                        xy=(optimal_k, optimal_score), 
                        xytext=(10, 10), textcoords='offset points',
                        bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7))
        
        # 3. Cluster distribution
        ax3 = axes[1, 0]
        cluster_counts = pd.Series(self.cluster_labels).value_counts().sort_index()
        bars = ax3.bar(cluster_counts.index, cluster_counts.values, alpha=0.7, color='skyblue', edgecolor='navy')
        ax3.set_title('📊 Questions per Cluster')
        ax3.set_xlabel('Cluster ID')
        ax3.set_ylabel('Number of Questions')
        
        # Add count labels on bars
        for bar in bars:
            height = bar.get_height()
            ax3.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                    f'{int(height)}', ha='center', va='bottom')
        
        # 4. Category distribution per cluster
        ax4 = axes[1, 1]
        cluster_category_crosstab = pd.crosstab(
            self.df['Cluster_Label'], 
            self.df['Category']
        )
        
        im = ax4.imshow(cluster_category_crosstab.values, cmap='YlOrRd', aspect='auto')
        ax4.set_title('🎭 Bloom Taxonomy Distribution per Cluster')
        ax4.set_xlabel('Bloom Taxonomy Category')
        ax4.set_ylabel('Cluster ID')
        
        # Set ticks
        ax4.set_xticks(range(len(cluster_category_crosstab.columns)))
        ax4.set_xticklabels(cluster_category_crosstab.columns, rotation=45)
        ax4.set_yticks(range(len(cluster_category_crosstab.index)))
        ax4.set_yticklabels(cluster_category_crosstab.index)
        
        # Add text annotations
        for i in range(len(cluster_category_crosstab.index)):
            for j in range(len(cluster_category_crosstab.columns)):
                text = ax4.text(j, i, cluster_category_crosstab.iloc[i, j],
                               ha="center", va="center", color="black", fontweight='bold')
        
        plt.colorbar(im, ax=ax4, label='Count')
        
        plt.tight_layout()
        plt.show()
        
        return fig
    
    def analyze_cluster_themes(self, top_n_samples=3, top_n_words=10):
        """
        Analyze what makes questions in each cluster similar
        """
        print("\n" + "="*80)
        print("🔍 CLUSTER THEME ANALYSIS")
        print("="*80)
        
        # TF-IDF untuk mencari kata-kata kunci per cluster
        tfidf = TfidfVectorizer(
            max_features=1000,
            stop_words='english',
            ngram_range=(1, 3),  # unigram, bigram, trigram
            min_df=2,
            max_df=0.8
        )
        
        # Fit TF-IDF pada semua pertanyaan
        tfidf_matrix = tfidf.fit_transform(self.df['Questions'])
        feature_names = tfidf.get_feature_names_out()
        
        for cluster_id in sorted(self.df['Cluster_Label'].unique()):
            cluster_questions = self.df[self.df['Cluster_Label'] == cluster_id]
            cluster_size = len(cluster_questions)
            
            print(f"\n🎯 CLUSTER {cluster_id} ({cluster_size} questions)")
            print("-" * 50)
            
            # 1. Kategori BT dominan
            bt_distribution = cluster_questions['Category'].value_counts()
            dominant_bt = bt_distribution.index[0]
            bt_percentage = (bt_distribution.iloc[0] / cluster_size) * 100
            
            print(f"📊 Dominant Bloom Taxonomy: {dominant_bt} ({bt_percentage:.1f}%)")
            if len(bt_distribution) > 1:
                other_bts = ", ".join([f"{bt}({count})" for bt, count in bt_distribution.iloc[1:3].items()])
                print(f"   Other categories: {other_bts}")
            
            # 2. Kata kunci karakteristik cluster
            cluster_indices = cluster_questions.index
            cluster_tfidf = tfidf_matrix[cluster_indices]
            
            # Hitung rata-rata TF-IDF untuk cluster ini
            cluster_mean_tfidf = np.array(cluster_tfidf.mean(axis=0)).flatten()
            
            # Hitung rata-rata TF-IDF untuk semua cluster lain
            other_indices = self.df[self.df['Cluster_Label'] != cluster_id].index
            other_tfidf = tfidf_matrix[other_indices]
            other_mean_tfidf = np.array(other_tfidf.mean(axis=0)).flatten()
            
            # Hitung selisih (distinctive words)
            tfidf_diff = cluster_mean_tfidf - other_mean_tfidf
            
            # Ambil kata-kata paling karakteristik
            top_word_indices = np.argsort(tfidf_diff)[-top_n_words:][::-1]
            characteristic_words = []
            
            for idx in top_word_indices:
                word = feature_names[idx]
                score = tfidf_diff[idx]
                if score > 0:  # Hanya kata yang benar-benar distinctive
                    characteristic_words.append(f"{word}({score:.3f})")
            
            print(f"🔑 Characteristic Keywords: {', '.join(characteristic_words[:8])}")
            
            # 3. Sample pertanyaan representatif
            print(f"📝 Representative Questions:")
            
            # Ambil pertanyaan yang paling dekat dengan centroid cluster
            if hasattr(self, 'embeddings'):
                cluster_embeddings = self.embeddings[cluster_indices]
                cluster_centroid = np.mean(cluster_embeddings, axis=0)
                
                # Hitung jarak ke centroid
                distances = []
                for i, embedding in enumerate(cluster_embeddings):
                    distance = cosine_similarity([embedding], [cluster_centroid])[0][0]
                    distances.append((distance, cluster_indices[i]))
                
                # Ambil yang paling representatif
                distances.sort(reverse=True)  # Cosine similarity: higher is better
                representative_indices = [idx for _, idx in distances[:top_n_samples]]
                
                for i, idx in enumerate(representative_indices, 1):
                    question = self.df.loc[idx, 'Questions']
                    category = self.df.loc[idx, 'Category']
                    # Truncate jika terlalu panjang
                    if len(question) > 120:
                        question = question[:120] + "..."
                    print(f"   {i}. [{category}] {question}")
            
            # 4. Semantic analysis
            self._analyze_semantic_patterns(cluster_questions)
            
            print()

    def _analyze_semantic_patterns(self, cluster_questions):
        """
        Analyze semantic patterns in cluster questions
        """
        questions = cluster_questions['Questions'].tolist()
        
        # Analisis pola semantic
        patterns = {
            'question_types': [],
            'common_themes': [],
            'cognitive_patterns': []
        }
        
        # Deteksi tipe pertanyaan
        question_words = ['what', 'how', 'why', 'when', 'where', 'which', 'who']
        for q_word in question_words:
            count = sum(1 for q in questions if q_word in q.lower())
            if count > 0:
                percentage = (count / len(questions)) * 100
                if percentage >= 20:  # Minimum 20% untuk dianggap signifikan
                    patterns['question_types'].append(f"{q_word.upper()}({percentage:.0f}%)")
        
        # Deteksi kata kerja kognitif (berdasarkan Bloom Taxonomy)
        cognitive_verbs = {
            'Remember': ['list', 'identify', 'name', 'state', 'define', 'recall', 'recognize'],
            'Understand': ['explain', 'describe', 'interpret', 'summarize', 'classify', 'compare'],
            'Apply': ['calculate', 'solve', 'demonstrate', 'apply', 'use', 'implement'],
            'Analyze': ['analyze', 'examine', 'compare', 'contrast', 'distinguish', 'differentiate'],
            'Evaluate': ['evaluate', 'assess', 'judge', 'criticize', 'justify', 'argue'],
            'Create': ['design', 'develop', 'create', 'formulate', 'plan', 'construct']
        }
        
        for level, verbs in cognitive_verbs.items():
            verb_count = 0
            for verb in verbs:
                verb_count += sum(1 for q in questions if verb in q.lower())
            
            if verb_count > 0:
                percentage = (verb_count / len(questions)) * 100
                if percentage >= 15:
                    patterns['cognitive_patterns'].append(f"{level}({percentage:.0f}%)")
        
        # Print patterns
        if patterns['question_types']:
            print(f"❓ Question Types: {', '.join(patterns['question_types'])}")
        
        if patterns['cognitive_patterns']:
            print(f"🧠 Cognitive Patterns: {', '.join(patterns['cognitive_patterns'])}")
        
        # Analisis panjang pertanyaan
        lengths = [len(q.split()) for q in questions]
        avg_length = np.mean(lengths)
        print(f"📏 Avg Question Length: {avg_length:.1f} words")
    
    def generate_cluster_insights_report(self):
        """
        Generate comprehensive insights about what makes clusters similar
        """
        print("\n" + "="*80)
        print("🎭 CLUSTER SIMILARITY INSIGHTS")
        print("="*80)
        
        insights = []
        
        for cluster_id in sorted(self.df['Cluster_Label'].unique()):
            cluster_questions = self.df[self.df['Cluster_Label'] == cluster_id]
            cluster_size = len(cluster_questions)
            
            # Analisis kategori BT
            bt_dist = cluster_questions['Category'].value_counts()
            dominant_bt = bt_dist.index[0]
            bt_dominance = (bt_dist.iloc[0] / cluster_size) * 100
            
            # Insight based on BT dominance
            if bt_dominance >= 70:
                bt_insight = f"Strong {dominant_bt} focus ({bt_dominance:.0f}%)"
            elif bt_dominance >= 50:
                bt_insight = f"Moderate {dominant_bt} emphasis ({bt_dominance:.0f}%)"
            else:
                bt_insight = f"Mixed cognitive levels (no clear dominance)"
            
            # Analisis panjang pertanyaan
            lengths = [len(q.split()) for q in cluster_questions['Questions']]
            avg_length = np.mean(lengths)
            
            if avg_length > 20:
                length_insight = "Complex, detailed questions"
            elif avg_length > 10:
                length_insight = "Medium-length questions"
            else:
                length_insight = "Short, concise questions"
            
            insights.append({
                'cluster_id': cluster_id,
                'size': cluster_size,
                'bt_insight': bt_insight,
                'length_insight': length_insight,
                'dominant_bt': dominant_bt
            })
            
            print(f"🔍 Cluster {cluster_id}: {bt_insight} + {length_insight}")
        
        # Summary insights
        print(f"\n📋 SUMMARY INSIGHTS:")
        print("-" * 40)
        
        # BT distribution across clusters
        bt_cluster_map = {}
        for insight in insights:
            bt = insight['dominant_bt']
            if bt not in bt_cluster_map:
                bt_cluster_map[bt] = []
            bt_cluster_map[bt].append(insight['cluster_id'])
        
        for bt, clusters in bt_cluster_map.items():
            if len(clusters) > 1:
                print(f"📚 {bt} questions are spread across clusters {clusters} - indicates topic diversity within same cognitive level")
            else:
                print(f"📖 {bt} questions are concentrated in cluster {clusters[0]} - indicates clear topic+cognitive grouping")
        
        return insights
    
    def generate_cluster_summary(self):
        """
        Generate detailed summary of clustering results
        """
        print("\n" + "="*80)
        print("📋 CLUSTERING SUMMARY REPORT")
        print("="*80)
        
        # Basic statistics
        n_questions = len(self.df)
        n_clusters = len(np.unique(self.cluster_labels))
        
        print(f"📊 Total Questions: {n_questions}")
        print(f"🎯 Number of Clusters: {n_clusters}")
        
        if self.silhouette_scores:
            optimal_k = max(self.silhouette_scores, key=self.silhouette_scores.get)
            print(f"🏆 Optimal Silhouette Score: {self.silhouette_scores[optimal_k]:.4f} (k={optimal_k})")
        
        # Cluster size distribution
        print(f"\n📈 Cluster Size Distribution:")
        cluster_sizes = pd.Series(self.cluster_labels).value_counts().sort_index()
        for cluster_id, size in cluster_sizes.items():
            percentage = (size / n_questions) * 100
            print(f"   Cluster {cluster_id}: {size} questions ({percentage:.1f}%)")
        
        # Category distribution analysis
        print(f"\n🎭 Bloom Taxonomy Distribution Analysis:")
        cluster_category_analysis = self.df.groupby('Cluster_Label')['Category'].value_counts().unstack(fill_value=0)
        
        for cluster_id in sorted(self.df['Cluster_Label'].unique()):
            print(f"\n   📌 Cluster {cluster_id}:")
            cluster_data = cluster_category_analysis.loc[cluster_id]
            total_in_cluster = cluster_data.sum()
            
            for category in cluster_data.index:
                count = cluster_data[category]
                if count > 0:
                    percentage = (count / total_in_cluster) * 100
                    print(f"      {category}: {count} ({percentage:.1f}%)")
        
        print("="*80)
        
        # Tambahkan analisis tema cluster
        self.analyze_cluster_themes()
        
        # Tambahkan insights report
        self.generate_cluster_insights_report()
        """
        Generate detailed summary of clustering results
        """
        print("\n" + "="*80)
        print("📋 CLUSTERING SUMMARY REPORT")
        print("="*80)
        
        # Basic statistics
        n_questions = len(self.df)
        n_clusters = len(np.unique(self.cluster_labels))
        
        print(f"📊 Total Questions: {n_questions}")
        print(f"🎯 Number of Clusters: {n_clusters}")
        
        if self.silhouette_scores:
            optimal_k = max(self.silhouette_scores, key=self.silhouette_scores.get)
            print(f"🏆 Optimal Silhouette Score: {self.silhouette_scores[optimal_k]:.4f} (k={optimal_k})")
        
        # Cluster size distribution
        print(f"\n📈 Cluster Size Distribution:")
        cluster_sizes = pd.Series(self.cluster_labels).value_counts().sort_index()
        for cluster_id, size in cluster_sizes.items():
            percentage = (size / n_questions) * 100
            print(f"   Cluster {cluster_id}: {size} questions ({percentage:.1f}%)")
        
        # Category distribution analysis
        print(f"\n🎭 Bloom Taxonomy Distribution Analysis:")
        cluster_category_analysis = self.df.groupby('Cluster_Label')['Category'].value_counts().unstack(fill_value=0)
        
        for cluster_id in sorted(self.df['Cluster_Label'].unique()):
            print(f"\n   📌 Cluster {cluster_id}:")
            cluster_data = cluster_category_analysis.loc[cluster_id]
            total_in_cluster = cluster_data.sum()
            
            for category in cluster_data.index:
                count = cluster_data[category]
                if count > 0:
                    percentage = (count / total_in_cluster) * 100
                    print(f"      {category}: {count} ({percentage:.1f}%)")
        
        print("="*80)
    
    def save_results(self, output_path='clustered_questions.csv'):
        """
        Save results to CSV file
        """
        print(f"💾 Saving results to {output_path}...")
        
        # Select only required columns
        output_df = self.df[['Questions', 'Category', 'Cluster_Label']].copy()
        
        # Save to CSV
        output_df.to_csv(output_path, index=False)
        print(f"✅ Results saved successfully!")
        
        return output_df

# Main execution function
def main(csv_path, output_path='clustered_questions.csv', clustering_method='kmeans'):
    """
    Main function to run the complete clustering pipeline
    
    Parameters:
    - csv_path: path to input CSV file
    - output_path: path for output CSV file
    - clustering_method: 'kmeans' or 'hdbscan'
    """
    
    print("🚀 Starting Semantic Question Clustering Pipeline")
    print("="*60)
    
    # Initialize clustering system
    clustering_system = SemanticQuestionClustering()
    
    try:
        # Load and preprocess data
        clustering_system.load_and_preprocess_data(csv_path)
        
        # Generate BERT embeddings
        clustering_system.generate_embeddings()
        
        # Reduce dimensions for visualization
        clustering_system.reduce_dimensions()
        
        # Perform clustering
        clustering_system.perform_clustering(method=clustering_method)
        
        # Create visualizations
        clustering_system.visualize_results()
        
        # Generate summary report
        clustering_system.generate_cluster_summary()
        
        # Save results
        final_results = clustering_system.save_results(output_path)
        
        print(f"\n🎉 Pipeline completed successfully!")
        print(f"📁 Results saved to: {output_path}")
        
        return clustering_system, final_results
        
    except Exception as e:
        print(f"❌ Error during processing: {str(e)}")
        raise e

# Example usage
if __name__ == "__main__":
    # Example usage - replace with your CSV file path
    CSV_FILE_PATH = "/kaggle/input/blooms-taxonomy-dataset/blooms_taxonomy_dataset.csv"  # Change this to your CSV file path
    OUTPUT_FILE_PATH = "clustered_questions_results.csv"
    
    print("🔧 Configuration:")
    print(f"   Input File: {CSV_FILE_PATH}")
    print(f"   Output File: {OUTPUT_FILE_PATH}")
    print(f"   Clustering Method: KMeans (with optimal k selection)")
    print()
    
    # Run the complete pipeline
    try:
        clustering_system, results = main(
            csv_path=CSV_FILE_PATH,
            output_path=OUTPUT_FILE_PATH,
            clustering_method='kmeans'  # or 'hdbscan'
        )
        
        print("\n✨ You can now:")
        print("1. Check the visualization plots")
        print("2. Review the clustering summary report") 
        print(f"3. Use the results from: {OUTPUT_FILE_PATH}")
        print("4. Access clustering_system object for further analysis")
        
    except FileNotFoundError:
        print(f"❌ Error: File '{CSV_FILE_PATH}' not found.")
        print("📝 Please:")
        print("1. Make sure your CSV file exists")
        print("2. Update CSV_FILE_PATH variable with correct path")
        print("3. Ensure CSV has 'Questions' and 'Category' columns")
```
